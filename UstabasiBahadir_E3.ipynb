```python
import random

def write_data(n, a, b, filepath):
    file = open(filepath, "w")

    for counter in range(0, n):
        string = str(random.randint(a,b)) + " "
        file.write(string)

def read_data(filepath):
    file = open(filepath, "r")
    data = file.read().split()
    return data

def sum_problem(filepath1, filepath2, k):
    list1 = read_data(filepath1)
    list2 = read_data(filepath2)

    for num1 in list1:
        for num2 in list2:
            if(int(num1) + int(num2) == k):
                print(num1, "+", num2, "=", k)

write_data(2000, 1, 10000,"C:\\Users\\oxika\\Desktop\\a1.txt")
write_data(2000, 1, 10000,"C:\\Users\\oxika\\Desktop\\a2.txt")
sum_problem("C:\\Users\\oxika\\Desktop\\a1.txt", "C:\\Users\\oxika\\Desktop\\a2.txt", 5000)

```

## 8.1 Rational numbers

```python
import math


def lowest(den3, num3):
    common_factor = math.gcd(int(num3), int(den3));

    den3 = int(den3 / common_factor);
    num3 = int(num3 / common_factor);
    return Rational(num3, den3)

def addFraction(num1, den1, num2, den2):
    den3 = math.gcd(den1, den2);

    den3 = (den1 * den2) / den3;

    num3 = ((num1) * (den3 / den1) +
            (num2) * (den3 / den2));

    return lowest(den3, num3);

def substractFraction(num1, den1, num2, den2):
    den3 = math.gcd(den1, den2);

    den3 = (den1 * den2) / den3;

    num3 = ((num1) * (den3 / den1) -
            (num2) * (den3 / den2));

    return lowest(den3, num3);

def divideFractions(o1, o2):
    return Rational(o1.p * o2.q, o1.q * o2.p)

class Rational:
    def __init__(self, p, q):
        divisor = math.gcd(p, q)
        p = p/divisor
        q = q/divisor

        self.p = int(p)
        self.q = int(q)

    def __str__(self):
        return f'{self.p}/{self.q}'

    def __add__(self, other):
        return addFraction(self.p, self.q, other.p, other.q)

    def __sub__(self, other):
        return substractFraction(self.p, self.q, other.p, other.q)

    def __truediv__(self, other):
        return divideFractions(self, other)

    def __eq__(self, other):
        return self.p == other.p and self.q == other.q

    def __float__(self):
        return float(self.p/self.q)



o1 = Rational(3, 1)
o2 = Rational(2, 5)


print(o1+o2)
print(o1-o2)
print(o1/o2)
print(o1==o2)
print(float(o2))
```

## 8.5 Implementing the set class

```python
class mySet:
    def __init__(self):
        self.values = {}

    def add(self, element):
        if element in self.values:
            return
        else:
            self.values[element] = 1

    def contains(self, element):
        if element in self.values:
            return True
        else:
            return False

    def remove(self, element):
        if element in self.values:
            self.values.pop(element)
        else:
            return False

    def __len__(self):
        return len(self.values)

    def __str__(self):
        return str(self.values)
```

## 8.6 Binary search tree

```python
class Node:
    def __init__(self, val=None):
        self.left = None
        self.right = None
        self.val = val

    def insert(self, val):
        if not self.val:
            self.val = val
            return

        if self.val == val:
            return

        if val < self.val:
            if self.left:
                self.left.insert(val)
                return
            self.left = Node(val)
            return

        if self.right:
            self.right.insert(val)
            return
        self.right = Node(val)

    def delete(self, val):
        if self == None:
            return self
        if val < self.val:
            self.left = self.left.delete(val)
            return self
        if val > self.val:
            self.right = self.right.delete(val)
            return self
        if self.right == None:
            return self.left
        if self.left == None:
            return self.right
        min_larger_node = self.right
        while min_larger_node.left:
            min_larger_node = min_larger_node.left
        self.val = min_larger_node.val
        self.right = self.right.delete(min_larger_node.val)
        return self

    def contains(self, val):
        if val == self.val:
            return True

        if val < self.val:
            if self.left == None:
                return False
            return self.left.contains(val)

        if self.right == None:
            return False
        return self.right.contains(val)

    def printTree(self, vals):
        if self.left is not None:
            self.left.printTree(vals)
        if self.val is not None:
            vals.append(self.val)
        if self.right is not None:
            self.right.printTree(vals)
        return vals

values = [21, 2, 0, 10, 24, 6, 18, 15, 11, 25, 13, 23]
bst = Node()

for val in values:
    bst.insert(val)

print(bst.printTree([]))
print(bst.contains(21))
print(bst.contains(19))

bst.insert(5)
bst.insert(7)

print(bst.printTree([]))

bst.delete(15)
bst.delete(11)

print(bst.printTree([]))
```
